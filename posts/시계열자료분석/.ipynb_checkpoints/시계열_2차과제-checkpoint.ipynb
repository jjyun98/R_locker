## 5번

```r
set.seed(2022)
et <- rnorm(100)
```

```r
zt <- rep(0, 100)
for(i in 1:99){
    zt[1] = 10
    zt[i+1] = 1 + 0.9*zt[i] + et[i]
    }
```

```r
df <- tibble(
    t = 1:100,
    zt = zt
    )
```

### 1)

```r
ggplot(data = df, aes(x = t, y = zt)) +
geom_point(color = 'steelblue') + 
geom_line(color = 'steelblue') +
theme_bw()
```

### 2)

```r
df %>% head
```

```r
acf(df[,2], lag.max = 10, main = "SACF (h = 1:10)")
```

- 시차가 늘어남에 따라 ACF는 감소하지만 결과적으로 시차 10까지 모두 자기상관관계가 존재함을 보여준다.<br>
결과적으로 시차 10까지는 영향을 미침.

### 3)

```r
pacf(df[,2], lag.max = 10, main = "SPACF (h = 1:10)")
```

- lag가 1인경우만 자기상관이 존재하고 그 이후로는 그렇지 않음을 볼 수 있다.<br>
즉, 한 시차전까지만 영향을 미친다는 의미이다.

### 4)

```r
df %>% mutate(zt_1 = c(df$zt[2:100],NA),zt_2 = c(df$zt[3:100],NA,NA)) -> df.2
df.2 %>% tail
```

```r
acf(df[,2], lag.max = 10, main = "SACF (h = 1:10)", plot = FALSE)[c(1,2)]
```

- 설명 : $\hat{\rho_1}$ = 0.932이고, $\hat{\rho_2}$ = 0.849이므로 $\hat{\rho_1}$에 비해 $\hat{\rho_2}$는 감소함.<br>
결국 $\rho$라는 것이 cov에서 나온 것이므로 선형관계가 약해진 것을 의미한다.<br>
먼저 직전 시차와 양의 부호 관계 였기에 양의 선형관계 모양으로 나온다.<br>
아래 두 그림을 보면 첫 번째에 비해 두 번째가 선형관계가 옅어진 것을 볼 수 있음.<br>
방향성 자체가 달라지지는 않으나 분산 자체가 늘어난 모습을 볼 수 있다.

```r
ggplot(data = df.2, aes(x = zt, y = zt_1)) +
geom_point() + 
ggtitle("Z_t, Z_t-1") + 
geom_hline(yintercept = 10.5, linetype='dotted', color='red', size=1) +
geom_vline(xintercept= 10.5 , linetype = 'dotted', color='red', size = 1)
```

### 5)

```r
ggplot(data = df.2, aes(x = zt, y = zt_2)) +
geom_point() + 
ggtitle("Z_t, Z_t-2") +
geom_hline(yintercept = 10.5, linetype='dotted', color='red', size=1) +
geom_vline(xintercept= 10.5 , linetype = 'dotted', color='red', size = 1)
```
