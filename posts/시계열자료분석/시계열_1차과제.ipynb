### 1)

- 선형 추세모형 식<br>
$z_t = \beta_0 + \beta_1 * t + \epsilon_t$

### 2)

```r
ltm <- lm(zt ~ t, data = ex2_data)
ltm %>% summary
```

추정된 회귀식은 $\beta_0 = 297.781, \beta_1 = 2.386$인 직선입니다.<br>
coef를 보았을 때, $\beta_0, \beta_1$(t로 표시된) 그리고 모형 자체 모두 p-value가 매우 작기에 유의하다고 볼 수 있고,<br>
R-squared가 0.8662인 것으로 보아 회귀모형에 대한 설명력 역시 높음을 의미합니다.

```r
ltm %>% dwtest
```

추세분석이기에 잔차분석 역시 보이면,<br>
DW = 3.19로 4에 가까워 1차 음의 자기상관관계를 보인다고 나오지만 p-value가 0.99로 매우 높기에 신뢰할 수 있는 정보는 아닙니다.

```r
res_2 <- resid(ltm)
res_2 %>% plot()
abline(h = 0)
```

추가적인 확인을 위해 잔차를 찍어보면 분명 줄어드는 것 같은 형태를 띄긴 하기에 음의 자기 상관관계가 있는 것으로 보입니다.

### movingaverage

##### female

```r
femaletimeseries = ts(female)
```

```r
fem.dat <- data.table(female = female,
                      t = 1:length(female))
fem.dat[, sma3 := SMA(femaletimeseries, n = 3)]
fem.dat[, sma10 := SMA(femaletimeseries, n = 10)]
```

```r
fem.dat %>% tail
```

```r
melt.female <- melt(fem.dat, id = 't')

ggplot(melt.female, aes(t, value, col = variable, size = variable, lty = variable)) +
geom_line() +
theme_bw() +
labs(col = "") +
scale_linetype_manual(values = c('solid', 'twodash', 'dashed')) +
scale_color_manual(values = c('black', 'orange', 'steelblue')) +
scale_size_manual(values = c(0.2, 1.2, 1.2)) +
guides(lty = 'none', size = 'none') +
theme(axis.title.x = element_blank())
```

```r
mean((fem.dat$female- fem.dat$sma3)^2, na.rm=T) ##MSE
mean(abs(fem.dat$female- fem.dat$sma3), na.rm=T) ##MAE
mean(abs((fem.dat$female- fem.dat$sma3)/fem.dat$female), na.rm=T)*100 ##MAPE

mean((fem.dat$female- fem.dat$sma10)^2, na.rm=T) ##MSE
mean(abs(fem.dat$female- fem.dat$sma10), na.rm=T) ##MAE
mean(abs((fem.dat$female- fem.dat$sma10)/fem.dat$female), na.rm=T)*100 ##MAPE
```

### 단순 지수 평활법

##### ses

##### alpha = 0.9 case

```r
fem.dat[, ses_0.2 := ses(female)$fitted]
```

```r
fem_ses_fit <- ses(female, 
             alpha = 0.9, # 사실 알파 안써도 얘가 적당히 넣어주긴함
             initial = 'simple',
             h = 10) # 앞으로 몇 개 더 예측해줄지 개수 설정
ls(fem_ses_fit)
```

```r
summary(fem_ses_fit)
```

```r
w <-c(seq(0.1,0.8,0.1), 
      seq(0.81, 0.99, 0.01)) # weight

sse <- sapply(w, function(x) 
  return(sum(ses(female, alpha = x)$residuals^2)))
              
sse %>% head
```

```r
w[which.min(sse)]  # 최적 평활상수값
```

0.99가 나오기는 하는데 0.99로 돌리면 예측의 의미가 너무 희석되는것 같기에 alpha = 0.9로 사용합니다.

`-` 평활 상수값에 따른 SSE plot

```r
w1 = w[-c(1:6)]  # xaxis from 0.7 to 1.0
sse1 = sse[-c(1:6)]
plot(w1,sse1, type="o", xlab="weight", ylab="sse", pch=16)
```

그냥 쭉 내려오기만 함.<br>
그래서 default에서 0.99로 잡아주는 것 같음.

- 잔차

```r
t.test(resid(fem_ses_fit), mu=0)
```

해당 ses의 잔차 출력한 것에 평균0인지 t.test검증 -> 유의

```r
plot(fem_ses_fit, xlab="", ylab="", 
     main="alpha=0.9", 
     lty=1,col="black" )
lines(fitted(fem_ses_fit), col="red", lty=2)
legend("topright", legend=c("female", "alpha=0.9"), 
       lty=1:2,col=c("black","red"))
```

alpha = 0.9라는 것은 실제 정보를 9만큼 예측값을 1만큼 반영(내분)하는 의미이기에<br>
파란색 줄은 너무 예측이 아닌 마지막 실제 값에 의존하는 모습을 보임.

`-` 잔차 그림

```r
plot(fem_ses_fit$residuals, ylab="residual",
     main="Time Series Plot of Prediction Errors : alpha=0.9"); abline(h=0)
```

```r
dwtest(lm(fem_ses_fit$residuals~1), alternative = 'two.sided')
```

보기에는 양수쪽으로 몰려있는 느낌도 있는데 더빈왓슨 결과 p-value 0.5이하에 DW = 2근처로 유의하다고 나옴.

0.2인 이유는 따로 없습니다. 0.9에 비해 예측값을 좀 더 많이 반영하기를 바라는 경우라서 했습니다.

### 이중지수 평활법

#### holt

```r
fem_holt_fit = holt(female, alpha=0.6, beta=0.6, h=12) 
summary(fem_holt_fit)
```

- 해석 : 계절별 예측을 보면 계절이 뒤로 갈 수록 늘어남. 근데 이게 계절 탓인 것은 아닌거 같고 그저 뒤로 갈 수록 증가하는 추세이기에 발생하는 모습으로 보임.

```r
book_holt_fit = holt(book, alpha=0.6, beta=0.6, h=7) 
summary(book_holt_fit)
```

```r
plot(fem_holt_fit, ylab="", xlab="",  lty=1, col="black",
     main="double smoothing"
     )
lines(fitted(fem_holt_fit), col="red", lty=2)
legend("topleft", lty=1:2, col=c("black","red"), c("female", "Double"), bty = "n")
```

```r
plot(book_holt_fit, ylab="", xlab="",  lty=1, col="black",
     main="double smoothing"
     )
lines(fitted(fem_holt_fit), col="red", lty=2)
legend("topleft", lty=1:2, col=c("black","red"), c("book", "Double"), bty = "n")
```

- 잔차 분석

```r
plot(resid(fem_holt_fit), main="Time series plot of forecasting error"); abline(h=0)
```

```r
plot(resid(book_holt_fit), main="Time series plot of forecasting error"); abline(h=0)
```

```r
dwtest(lm(resid(fem_holt_fit)~1), alternative = 'two.sided')
```

```r
dwtest(lm(resid(book_holt_fit)~1), alternative = 'two.sided')
```

book의 p-value가 높긴하지만 둘 다 2근처에 이 정도면 유의하다고 볼 수 있음.<br>
그림도 이상없음.

### 계절평활법
